```python
# Find largest number in row
df_max = testdf.max(axis=1)
```

```python
# devide numbers with this number, rounded, and replace
for index in list(range(0,len(df_max),1)):
    testdf.iloc[index][barges] = testdf.iloc[index][barges] / df_max[index] * 10
```

```python
df_new = testdf.loc[:,barges].div(df_max, axis=0).round(1) * 10
```
